```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
iris.target
```

```python
df["target"] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df["target_name"] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df["target"]==0]
df0
```

```python
df1 = df[df["target"]==1]
df1
```

```python
df2 = df[df["target"]==2]
df2
```

```python
df.columns
```

```python
%matplotlib inline
```

```python
plt.title("sepal length and width of setosa and versicolor", fontsize=14)
plt.xlabel("sepal length (cm)", fontsize=14)
plt.ylabel("sepal width (cm)", fontsize=14)
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], color="r", marker="+")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color="g", marker=".")
plt.show()
```

```python
plt.title("petal length and width of setosa and versicolor", fontsize=14)
plt.xlabel("petal length (cm)", fontsize=14)
plt.ylabel("petal width (cm)", fontsize=14)
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color="purple", marker="+")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color="blue", marker=".")
plt.show()
```

```python
x = df.drop(["target", "target_name"], axis=1)
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```
